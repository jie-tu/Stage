```python
import pandas as pd
```

```python
data  = 'G:/516/Int516.csv'
```

```python
df = pd.read_csv(data)
df.sample(5)
```

```python
df.drop(df[df.hubDist >100].index, inplace=True)
df.drop(df[df.modes ==11].index, inplace=True)
```

(df.corr().loc['modes'].plot(kind='barh', figsize= (4,14)))

```python
corr = abs(df.corr().loc['modes'])
corr = corr[corr<0.2]
cols_to_drop = corr.index.to_list()
df = df.drop(cols_to_drop, axis=1)
```

```python
df.sample(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(16,10)})
sns.heatmap(df.corr(),
annot=True,
linewidths=.5,
center=0,
cbar=False,
cmap="YlGnBu")
plt.show()
```
