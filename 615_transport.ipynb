## load file 

```python
file = open("Paris", encoding='utf-8')
result = api.parse_xml(file.read(), encoding='utf-8', parser=None)
```

## get all bus line

```python
relation_bus= {}
for relation in result.relations:
    for member in relation.members:
        if(member.role=='stop'):
            if 'name' in relation.tags.keys():
                if 'Bus' in relation.tags['name']:
                    relation_bus [relation.id]= relation.tags['name']
```

## get relations from bus id

```python
bus_to_relation= {}
for relation in result.relations:
    for member in relation.members:
        if 'name' in relation.tags.keys():
            if 'Bus' in relation.tags['name']:
                if(member.role=='stop'):
                    if member.ref in bus_to_relation :
                        bus_to_relation [member.ref].append(relation.id)
                    else:
                        bus_to_relation [member.ref] = [relation.id]
                    
```
